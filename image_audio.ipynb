```python
import pytesseract
from PIL import Image
import pygame
import gtts
```

```python
img = Image.open('/content/image_to_text_sample3.png')
print(img)
```

```python
text = pytesseract.image_to_string(img)
print(text)
tts = gtts.gTTS(text)
tts.save('audiofile.mp3')
```

```python
from IPython.display import Audio

audio_path = "/content/audiofile.mp3"

# Play the audio file
Audio(audio_path)
```
